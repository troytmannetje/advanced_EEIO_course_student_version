```python
# import modules
import pandas as pd
import numpy as np

# Import data
supply_ = pd.read_excel("data/modified_supply_basic_price_AT_OECD_2015.xlsx", engine="openpyxl", skipfooter=3)
print(supply_.index)
```

```python
# organize you row and column labels
supply_col_lb = supply_.iloc[9,7:].str.strip().tolist()
supply_row_lb = supply_.iloc[11:, 2].str.strip().tolist()

# isolate array with only values
supply = supply_.iloc[11:,7:]

# assign clean labels to columns and row
supply.columns = supply_col_lb
supply.index = supply_row_lb

#print(*supply.columns, sep="\n")

supply
```

## Exercise 1

1. Look at the product list and select two products

```python
# Print product list (i.e., the index of your table)
print(*supply.index, sep="\n")
```

```python
# Select two products
products = ["Basic metals", "Electrical equipment"]
```

2. Can you quantify their total supply, import, and which industries supply them?

```python
# Slice your data
# replace none with your function
p_1_supply = supply.loc[products[0],:]
p_2_supply = supply.loc[products[1],:]

# Isolate non zero values 
p_1_supply = p_1_supply[p_1_supply>0].sort_values(ascending=False)
p_2_supply = p_2_supply[p_2_supply>0].sort_values(ascending=False)

# Print results
print(f"{products[0]}:\n{p_1_supply}\n\n{products[1]}:\n{p_2_supply}")
```

3. What is the total product output at basic price and at purchaser price?

    N.b. in the Supply and Use table Non-Deductible VAT is included in the Taxes less subsidies on products (see also page 86 Eurostat Manual of Supply, Use and Input-Output Tables)

Tip: in the file "modified_supply_basic_price_AT_OECD_2015.xlsx" you can check whether the calculations are right in the columns "Total supply at purchasers' prices" and "Total supply at basic prices"

Tip: you can use p_1_supply.loc[] to select the right rows

```python
for product in products:  
    # Total supply at basic and purchaser prices for product
    p_purchaser_price = supply.loc[product,:].sum()
    
    p_basic_price = np.sum(p_purchaser_price-
                           supply.loc[product, "Trade and transport margins"]-
                           supply.loc[product, "Taxes less subsidies on products"])
    
    
    # basic and purchase prices for product
    print(f"{product}: \n\tBasic price: {p_basic_price:.0f}\n\tPurchaser price: {p_purchaser_price:.0f}\n\tdifference: {p_purchaser_price/p_basic_price*100:.0f}%\n")

```
